```python
import geopandas as gpd
goias_cities = gpd.read_file("GO_Municipios_2020/GO_Municipios_2020.shp")
goias_cities.plot()
```

```python
goias_cities.head()
```

```python
ifg_campuses = [
    #'Águas Lindas',
    'Anápolis',
    #'Aparecida de Goiânia',
    #'Cidade de Goiás', 
    'Formosa', 
    'Goiânia', 
    #'Goiânia Oeste', 
    'Inhumas', 
    #'Itumbiara', 
    'Jataí', 
    'Luziânia', 
    #'Senator Canedo', 
    'Uruaçu'
    #'Valparaíso'
]
ifg_campuses = goias_cities.loc[goias_cities['NM_MUN'].isin(ifg_campuses)].reset_index(drop=True)
ifg_campuses
```

```python
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
import matplotlib.patches as mpatch
from matplotlib.patches import FancyBboxPatch
from matplotlib.text import OffsetFrom

ifg_campuses['coords'] = ifg_campuses['geometry'].apply(lambda x: x.representative_point().coords[:])
ifg_campuses['coords'] = [coords[0] for coords in ifg_campuses['coords']]


fig, ax = plt.subplots(1, 1, figsize  = (20, 18))
ax.set_axis_off();

goias_plot = goias.plot(
    column = 'pdeig',
    cmap = 'tab10', 
    alpha=.5, 
    ax=ax, 
    legend=True)

ifg_campuses.plot(ax=goias_plot, alpha=.2, hatch="/////", legend=True);
for idx, row in ifg_campuses.iterrows():
    plt.annotate(
        text=row['NM_MUN'], 
        xy=row['coords'],
        horizontalalignment='center',
        fontsize=10,
        xycoords="data",
        va="top", 
        ha="center",
        bbox=dict(boxstyle="round", fc="w", alpha=0.8, edgecolor='red') 
    )
leg = ax.get_legend()
leg.set_bbox_to_anchor((0.1, .7, 0.2, 0.2))    
```
